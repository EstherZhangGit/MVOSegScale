```python
#coding:utf-8
import random
import numpy as np
import cv2
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 5)

img = cv2.imread('/Volumes/Matrix/Dataset/KITTI/00/image_2/000000.png',
                 cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img1 = img.copy()
img2 = img.copy()
img3 = img.copy()
img4 = img.copy()
img5 = img.copy()

plt.imshow(img)
plt.show()
```

```python
akaze = cv2.AKAZE_create(threshold=0.000001)
kp_akaze = akaze.detect(img, None)
cv2.drawKeypoints(img1,
                  kp_akaze,
                  img1,
                  flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img1)
plt.plot()
print(len(kp_akaze))
```

```python
orb = cv2.ORB_create(nfeatures=8000, scaleFactor=1.1, nlevels=20)
kp_orb = orb.detect(img, None)
cv2.drawKeypoints(img2,
                  kp_orb,
                  img2,
                  flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img2)
plt.plot()
print(len(kp_orb))
```

```python
brisk = cv2.BRISK_create(thresh=15)
kp_brisk = brisk.detect(img, None)
cv2.drawKeypoints(img3,
                  kp_brisk,
                  img3,
                  flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img3)
plt.plot()
print(len(kp_brisk))
```

```python
mser = cv2.MSER_create()
kp_mser = mser.detect(img, None)
cv2.drawKeypoints(img4,
                  kp_mser,
                  img4,
                  flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img4)
plt.plot()
print(len(kp_mser))
```

```python
sift = cv2.SIFT_create()
kp_sift = sift.detect(img, None)
cv2.drawKeypoints(img5,
                  kp_sift,
                  img5,
                  flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img5)
plt.plot()
print(len(kp_sift))
```
